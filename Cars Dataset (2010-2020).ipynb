## Overview

## Dataset Description

The dataset includes essential attributes such as:

- **Make and Model**: The brand and specific model of the car.
- **Year of Manufacture**: The year the car was manufactured.
- **Engine Size**: The size of the car’s engine.
- **Fuel Type**: The type of fuel the car uses (e.g., petrol, diesel, electric).
- **Price (USD)**: The price of the car in US dollars.

These attributes allow for in-depth analysis and modeling, facilitating insights into market trends, price prediction, and comparative studies among different car makes and models.

## 2. Sales Trend by Fuel Type: What are the car sales trends over the years based on fuel type?

Example: How have the sales of petrol, diesel, electric, and hybrid cars changed from 2010 to 2020?

```python
data_year_fuel = pd.pivot_table(data,values = "Engine Size (L)", index = ['Fuel Type'], columns =['Year'], aggfunc = np.count_nonzero)
```

```python
data_year_fuel
```

```python
plt.figure(figsize=(10,3))
plt.plot(data_year_fuel.iloc[0,:],'ro-')
plt.title("Diesel Car Sell Trend")
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot(data_year_fuel.iloc[1,:],'bo-')
plt.title("Electric Car Sell Trend")
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot(data_year_fuel.iloc[2,:],'go-')
plt.title("Hybrid Car Sell Trend")
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot(data_year_fuel.iloc[3,:],'ro-')
plt.title("Petrol Car Sell Trend")
plt.show()
```

## 4. Best Selling Model per Company per Year: Which models were the best sellers for each company every year?
